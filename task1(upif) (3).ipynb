```python
# ============================================================
# QUESTION: Average daily transaction frequency?
# ============================================================


# ------------------------------------------------------------
#  Per-user daily frequency estimate
# using account age as a stand-in for "observation window in days".
# This gives a genuine average transactions-per-day-per-user metric,
# even without calendar dates.
# ------------------------------------------------------------
if 'receiver_account_age' in df.columns:
    user_activity = (
        df.groupby('user_id')
        .agg(transaction_count=('user_id', 'size'),
             account_age_days=('receiver_account_age', 'max'))
        .reset_index()
    )
    # avoid divide-by-zero for brand-new accounts (age = 0)
    user_activity['account_age_days'] = user_activity['account_age_days'].replace(0, 1)
    user_activity['avg_daily_frequency'] = (
        user_activity['transaction_count'] / user_activity['account_age_days']
    )

    overall_avg_daily_frequency = user_activity['avg_daily_frequency'].mean()

    print(f"\nEstimated average daily transaction frequency per user: "
          f"{overall_avg_daily_frequency:.4f} transactions/day")
    print("\nTop 10 users by estimated daily frequency:")
    print(user_activity.sort_values('avg_daily_frequency', ascending=False).head(10))
else:
    print("\n'receiver_account_age' column not found — cannot estimate per-user daily frequency.")
```
